# E-Commerce Sales Analysis using Pandas

This project explores an e-commerce dataset using Python and Pandas to answer business questions and practice data analysis skills.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('Display.max_columns',None)
```

```python
df = pd.read_csv('E-Commerce Sales.csv')
df
```

## Display last 10 Rows of the dataset

## Check data type of each column

```python
# Task 3: Check the type of columns
# To check the datatype of each column we use dtypes attribute
print('Datatype of each column in the dataset')
print('------------------------------------------')
print(df.dtypes)
```

## Check number of null values in each column

```python
# Task 4: Check number of null values in each column
# To check the number of null values in each column we use isna() and sum() functions
print('Null values in each column')
print('------------------------------------------------')
df.isna().sum()
```

## Find total number of rows and  columns in the Dataset

```python
# Task 5: Find total number of rows and  columns in the Dataset
# To find total number of rows and columns in the dataset we use shape attribute
print('Total number of Rows and columns in the Dataset')
print('-----------------------------------------------')
for rows,columns in [df.shape]:
    print('Number of rows =',rows)
    print('Number of columns =',columns)
```

## Find the highest and lowest PurchasePrices

```python
# Task 6: Find the highest and lowest PurchasePrices
# To find maximum and minimum purchase prices we use min() and max() functions
print('-----------------------------------------------------------------------')
print('Lowest Purchase Price')
print('-----------------------------------------------------------------------')
print(df[df['Purchase Price']==df['Purchase Price'].min()]['Purchase Price'])
print('-----------------------------------------------------------------------')
print('Highest Purchase Prices')
print('-----------------------------------------------------------------------')
print(df[df['Purchase Price']==df['Purchase Price'].max()]['Purchase Price'])
```

## Find the average purchase price

```python
# Task 7: Find the average purchase price
# To find the average we use mean() function
print('Average Purchase price =',df['Purchase Price'].mean())
```

## How many people using French 'fr' language

```python
# Task 8: How many people using French 'fr' language
# To find how many people using french language we access the data of language using French
# Then we find the length of the data 
print('Number of people using french as their language',len(df[df['Language']=='fr']))
```

## Find the JobTitle contains engineer

```python
# Task 9: Find the JobTitle contains engineer
# To find the jobtitle containing engineer we access the Job column then obtain the job titles with engineer
print('Job Titles containing engineer')
print('------------------------------------------')
print(df[df['Job'].str.contains('engineer')]['Job'])
```

## Find the email of the person with the IP Address: 132.207.160.22

```python
# Task 10: Find the email of the person with the IP Address: 132.207.160.22F
# To obtain the email of the IP address given we access the data of the IP Address then access the email column
print('Email of the IP Address:132.207.160.22')
print('-------------------------------------------------')
print(df[df['IP Address']=='132.207.160.22']['Email'])
```

## How many people has master card as their credit card provider and made a purchase above 50

```python
# Task 11: How many people has master card as their credit card provider and made a purchase above 50
# To find number of people using Master card and purchase prices above 50
# Convert the Purchase Price column to float  and then we use conditional statements to obtain number of people
print('Number of people using Mastercard and PurchasePrice is above 50 =',df[(df['CC Provider']=='Mastercard')&(df['Purchase Price'] > 50)].shape[0])
```

##  How many people purchase during the AM and how many people purchase during PM

```python
# Task 13: How many people purchase during the AM and how many people purchase during PM
# To find number purchases at AM and PM we use value_counts() function
print('Number of people purchasing during AM and PM')
print('-----------------------------------------------------')
df['AM or PM'].value_counts()
```

## How many people have a credit card that expires in 2020

```python
# Task 14: How many people have a credit card that expires in 2020
# To find number of people have a credit card that expires in 2020 should access the data of exp date column ending with 2020
# Then using shape attribute we can obtain count of people
print('Number of people credit card expires in 2020 =',df[df['CC Exp Date'].str.endswith('20')].shape[0])
```

## Find the Address of top 5 highest purchase prices

```python
# Task 15: Find the Address of top 5 highest purchase prices
# To find the address of top 5 highest purchase prices use nlargest() function and access the Address
df.nlargest(5,'Purchase Price')['Address']
```

## What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)

```python
# Task 16: What are the top 5 most popular email providers
# To find the top 5 most popular email providers --> first access the Email column
# Use split() to separate the email provider name then access the number of unique values using value_counts()
# Use head() to access top 5
df['Email'].str.split('@').str[1].value_counts().head()
```

# Visualization

## Visualize the bar graph for the Language counts

```python
# Task 17: Visualize the bar graph for the language counts
# To obtain a bar graph of languages find the language counts using value_counts
# Using plt.bar() obtain a bar graph
lang_counts = df['Language'].value_counts().values
lang = df['Language'].value_counts().index
plt.bar(lang,lang_counts)
plt.title('Language Distribution')
plt.xlabel('Language')
plt.ylabel('Count')
plt.show()
```

## Visualize a pie chart of Top 10 Lots with highest purchase price

```python
# Task 18: Visualize a horizontal bar graph of Top 10 Lots with highest purchase price
# To visualize top 10 lots with highest purchase prices sort purchase price using sort_values access top 10 lots
# Using plt.barh() obtain horizontal bar graph
ind = df['Purchase Price'].sort_values(ascending = False).head(10).index
prices = df['Purchase Price'].sort_values(ascending = False).head(10).values
lot = df.loc[ind,'Lot'].values
plt.barh(lot,prices)
plt.title('Top 10 Highest Purchase Prices')
plt.xlabel('Purchase Price')
plt.ylabel('Lot')
plt.show()
```

## Visualize the Purchase Price distribution

```python
# Task 19: Visualize the  Purchase Price distribution
# To visualize Purchase Price distribution access the Purchase Price column
# Using hist() obtain the histogram
plt.figure(figsize=(10,5))
plt.hist(df['Purchase Price'],bins=50)
plt.title('Purchase Price Distribution')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.show()
```

## Visualize the Top 5 Most Popular Email Providers

```python
# Task 20: Visualize the top 5 most popular email providers
# To visualize the top 5 most popular email providers access the top 5 email providers and obtain the value counts
# Using plt.pie() visualize top 5 email providers
email_counts = df['Email'].str.split('@').str[1].value_counts().head().values
emails = df['Email'].str.split('@').str[1].value_counts().head().index
exp = [0,0,0,1,1]
plt.pie(email_counts,labels = emails,autopct='%0.2f%%',explode = exp)
plt.show()
```
